## The code below is a practice to generate an image: 
- the value of each pixel in the i th image row of the red channel is  i*0.0039. 
- the value of each pixel in the i th image row of the green channel is  1-i*0.0039
- the value of each pixel in blue channel is 1

## Matplotlib's imshow: When you use plt.imshow(im), Matplotlib interprets the three channels as RGB values and displays the image accordingly.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 256
im = np.zeros((N, N, 3))
for i in range(0, N):
    im[i, :, 0] = i*0.0039
    im[i, :, 1] = 1 - i*0.0039
    im[i, :, 2] = 1

plt.imshow(im)
```
