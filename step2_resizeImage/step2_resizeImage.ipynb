```python
import pandas as pd
import numpy as np

from glob import glob
import os

import matplotlib.pyplot as plt
import cv2
```

```python
## For windows, use the following code to get all the .jpg files in a directory:
dog_files = glob(os.path.join('..', 'data', 'test_set', 'dogs', '*.jpg'))
cat_files = glob(os.path.join('..', 'data', 'test_set', 'cats', '*.jpg'))

## sort the files
dog_files.sort()
cat_files.sort()
```

## 1. Show the original image

```python
image_mpl = plt.imread(dog_files[5])

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(image_mpl) # Display an image on the axes
ax.axis('off') # Turn off the axis
plt.show()
```

#### 2. change the size with ratio

```python
image_resize = cv2.resize(image_mpl, None, fx=0.25, fy=0.25) # Resize the image to 25% of the original size

## Display the resized image:
####  as we fix the plot size, 
#### the image will be displayed with lower resolution
fig, ax = plt.subplots(figsize=(8, 8)) 
ax.imshow(image_resize)
ax.axis('off')

plt.show()
```

```python
## compare the original image with the resized image
print("Original image shape:", image_mpl.shape)
print("Resized image shape:", image_resize.shape)
```

#### 3. change the size with fix number

```python
image_resize_number = cv2.resize(image_mpl, (100, 200)) # Resize the image to 100x100 pixels

## Display the resized image:
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(image_resize_number)
ax.axis('off')

plt.show()
```

```python
image_resize_number.shape # Check the shape of the resized image
```

#### 4. resize the image with fix number, but we need to keep the original weight/hight ratio

```python
image_resize_number_interplation = cv2.resize(image_mpl, (5000, 5000), interpolation=cv2.INTER_CUBIC) # Resize the image to 100x200 pixels using the INTER_CUBIC interpolation method
## the INTER_CUBIC interpolation method is slower than the INTER_LINEAR method, but it produces higher quality images
## interpolation is a method used to estimate the pixel values of the resized image

## Display the resized image:
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(image_resize_number_interplation)
ax.axis('off')

plt.show()
```

```python
## Check the shape of the resized image
image_resize_number_interplation.shape
```
