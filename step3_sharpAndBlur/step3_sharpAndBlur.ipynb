```python
import pandas as pd
import numpy as np

from glob import glob

import matplotlib.pyplot as plt
import cv2
```

```python
## For windows, use the following code to get all the .jpg files in a directory:
dog_files = glob("data/test_set/dogs\\*.jpg")
cat_files = glob("data/test_set/cats\\*.jpg")
```

```python
image_mpl = plt.imread(cat_files[2])

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(image_mpl) # Display an image on the axes
ax.axis('off') # Turn off the axis
ax.set_title('original Image') # Set the title of the plot
plt.show()
```

## Use online-available kernel to sharpen the image 
- https://en.wikipedia.org/wiki/Kernel_(image_processing)

```python
## sharpen the image
kernel_sharp = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
```

```python
## Apply the kernel to the image using the filter2D function
## The -1 in the third argument specifies that the output image has the same depth as the input image
## The cv2.filter2D function convolves the kernel with the image

image_sharp = cv2.filter2D(image_mpl, -1, kernel_sharp)

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(image_sharp)
ax.set_title('Sharpened Image')
ax.axis('off')

plt.show()
```

## Use online-available kernel to bluring the image

```python
## blur the image
kernel_blur = np.array([[1,1,1], [1,1,1], [1,1,1]]) / 9
```

```python
image_blur = cv2.filter2D(image_mpl, -1, kernel_blur)

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(image_blur)
ax.set_title('Blurred Image')
ax.axis('off')

plt.show()
```
